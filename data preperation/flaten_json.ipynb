```python
import json

with open("data/procedures/osha_confined_spaces_faq.json", "r", encoding="utf-8") as f:
    data = json.load(f)

with open("data/procedures/osha_confined_spaces_faq.md", "w", encoding="utf-8") as out:
    for item in data:
        qnum = item.get("question_number", "")
        question = item.get("question", "No question")
        answer = item.get("answer", "No answer")

        out.write(f"## Q{qnum}. {question}\n\n")
        out.write(f"{answer}\n\n")
        out.write("---\n\n")
```
